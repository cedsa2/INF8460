```python
import pandas as pd

df = pd.read_csv("output/answers_test_tfidf_ngrams_15000_cosine_top25.csv")
df.head(20)
```

```python
df.loc[df.score <= 0.25, "answer"] = "<No Answer>"
df.drop(['score'], axis=1, inplace=True)
df.head()
```

```python
df.to_csv('output/submission_answers_test_tfidf_ngrams_15000_cosine_top25.csv', index=False)
```

```python
unfiltered = df.answer
filtered = []
for ans in unfiltered:
#     print(ans)
    ans = ans.split(',')
#     print(len(ans))
    if len(ans) > 2:
        text = ','.join(ans[0:len(ans)-1])
    else:
        text = ans[0]
#     print(text)
    score = float(ans[-1])
    if score < 8.0:
        filtered.append('<No Answer>')
    else:
        filtered.append(str(text))
# df.drop(['answer'], axis=1, inplace=True)
df['answer'] = filtered
df['answer'] = df['answer'].astype(str)
df.head(20)
```

```python
df.to_csv('output/submission_answers_val_electra_tfidf_15000_cosine_top25.csv', index=False)
```

```python
import pandas as pd

df = pd.read_csv("output/ranking_test_tfidf_15000_cosine_top25.csv")
df.head(20)
```

```python
ll = []
for l in df["top_25"]:
    l = l[1:len(l)-1]
    l = l.split(', ')
    ll.append(";".join(l))
    
df['top_n'] = ll
df.drop(['top_25'], axis=1, inplace=True)

ll = []
for l in df["top_10"]:
    l = l[1:len(l)-1]
    l = l.split(', ')
    ll.append(";".join(l))
    
df['top_10'] = ll
```
